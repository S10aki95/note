```python
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv("../../data/sample/winequality-red.csv", sep=";")
```

# Regression

$$Y = \beta^T X + \epsilon$$

$$Y = Poisson(\beta^T X + \epsilon)$$

```python
Y = df[["alcohol"]]
X = df[["density", "volatile acidity"]]
```

```python
X_b = sm.add_constant(X)
```

```python
model = sm.OLS(Y, X_b)
results = model.fit()
```

```python
print(results.summary())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X["density"], Y, color="red", label="Test Data")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression using statsmodels")
plt.legend()
plt.grid(True)
plt.show()
```
